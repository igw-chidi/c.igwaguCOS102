```python
# Equations Calculator
import numpy as np
print("What type of equation would you like to calculate")
print("1.Quartic Equation( (ax^4 + bx^3 + cx^2 + dx + e = 0)")
print("2.Cubic Equation( (ax^3 + bx^2 + cx + d = 0)")
print("3.Quadratic (ax^2 + bx + c = 0)")
choice = input("Enter your choice")

if choice == "1":
     a = float(input("Enter a"))
     b = float(input("Enter b"))
     c = float(input("Enter c"))
     d = float(input("Enter d"))
     e = float(input("Enter e"))
     coeff = [a,b,c,d,e]

elif choice == "2":
    a=float(input("Enter a"))
    b=float(input("Enter b"))
    c=float(input("Enter c"))
    d=float(input("Enter d"))
    coeff = [a,b,c,d] 
      
elif choice == "3":
    a = float(input("Enter a"))
    b = float(input("Enter b"))
    c = float(input("Enter c"))      
    coeff = [a,b,c]  
  
else: 
    print("Invalid choice") 
    return 


roots = np.roots(coeff)
print("The roots are:")
for i, root in enumerate(roots, 1):
    # formatting to show real and complex parts cleanly 
   print(f"root {i}: {root:.4f}")


    
    
        


 
```
